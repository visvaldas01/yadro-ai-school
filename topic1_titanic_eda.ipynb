**Let's create a new feature.**

```python
def age_category(age):
    """
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    """
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data["Age_category"] = age_categories
```

**Another way is to do it with `apply`.**

```python
data["Age_category"] = data["Age"].apply(age_category)
```

```python
# You code here
young = data[data['Age'] < 30]
young_survived = young[young['Survived'] == 1]
young_survived_share = round(100 * len(young_survived) / len(young), 1)
old = data[data['Age'] > 60]
old_survived = old[old['Survived'] == 1]
old_survived_share = round(100 * len(old_survived) / len(old), 1)
print(f"{young_survived_share}% among young and {old_survived_share}% among old")
```
