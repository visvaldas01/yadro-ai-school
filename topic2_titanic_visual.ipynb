**6. How does survival depend on passengers' age?  Verify (graphically) an assumption that youngsters (< 30 y.o.) survived more frequently than old people (> 55 y.o.).**

```python
# You code here
def age_category(age):
    """
    < 30 -> 1
    >= 30, <55 -> 2
    >= 55 -> 3
    """
    if age < 30:
        return 1
    elif age <= 55:
        return 2
    elif age > 55:
        return 3
train_df['Age_category'] = train_df['Age'].apply(age_category)
survival_rates = train_df.groupby('Age_category')['Survived'].mean().reset_index()
sns.barplot(x='Age_category', y='Survived', data=survival_rates)
```
